```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the sample climate dataset (replace 'your_data.csv' with your actual file)
df = pd.read_csv('monthly_data.csv')
df.head
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Data Cleaning and Preprocessing (replace this with your specific preprocessing steps)
# EDA
# Plot time series for minimum temperature and precipitation
plt.figure(figsize=(12, 6))
plt.plot(df['DATE'], df['MonthlyMinimumTemperature'], label='Minimum Temperature', color='orange')
plt.xlabel('Date')
plt.ylabel('Minimum Temperature (°C)')
plt.title('Minimum Temperature Over Time')
plt.legend()
plt.show()
```

```python
# Data Cleaning and Preprocessing (replace this with your specific preprocessing steps)
# EDA
# Plot time series for maximum temperature and precipitation
plt.figure(figsize=(12, 6))
plt.plot(df['DATE'], df['MonthlyMaximumTemperature'], label='Maximum Temperature', color='orange')
plt.xlabel('Date')
plt.ylabel('Maximum Temperature (°C)')
plt.title('Maximum Temperature Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['DATE'], df['MonthlyTotalLiquidPrecipitation'], label='Precipitation', color='blue')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.title('Precipitation Over Time')
plt.legend()
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = df[['MonthlyMinimumTemperature', 'MonthlyTotalLiquidPrecipitation']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Temporal Analysis
# Calculate moving averages for temperature
df['MonthlyMeanTemperature'] = df['MonthlyMinimumTemperature'].rolling(window=12).mean()

plt.figure(figsize=(12, 6))
plt.plot(df['DATE'], df['MonthlyMeanTemperature'], label='Temperature (Mean)', color='green')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Monthly Mean Temperature Over Time')
plt.legend()
plt.show()
```
